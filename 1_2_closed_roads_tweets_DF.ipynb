```python
tweetCriteria = got.manager.TweetCriteria().setQuerySearch(searchQuery)\
                                           .setMaxTweets(1000)
tweet = got.manager.TweetManager.getTweets(tweetCriteria)
#tweet[9].text
```

```python
for i in range(len(tweet)):
    print(tweet[i].text)
```

```python
import regex as re
text = []
for i in range(len(tweet)):
    text.append(re.sub(r'@\S+|https?://\S+', '', tweet[i].text))
text
```

"closed" doesn't seem to be necessary to return. it is before any mention of street in all tweets.

```python
tweet_lists = [i for i in text]
```

```python
roads_closed = []
for i in range(len(text)):
    roads_closed.append(re.findall(r'\d{1,3}', text[i]))
roads_closed
```

```python
roads = []
for i in roads_closed:
    k =[j for j in i if int(j) in [60, 28, 2, 38, 16, 3, 93, 99]]
    roads.append(k)
roads
```

```python
road_names = {'16': 'MA 16', '3': 'US 3', '93': 'I 93',
     '2': 'MA 2A', '60':'MA 60', '28': 'MA 28',
     '38':'MA 38','99':'MA 99'
    }
```

```python
closed_roads = [[road_names[x] for x in row] for row in roads]
closed_roads
```
